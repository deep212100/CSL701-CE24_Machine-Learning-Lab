Task 11: Convert Fare column to a NumPy array.

```python
fare_array = df['Fare'].to_numpy()
print(fare_array)
```

Task 12: Calculate maximum and average Fare using np.max() and np.mean().

```python
import numpy as np
max_fare = np.max(fare_array)
mean_fare = np.mean(fare_array)
print("Maximum Fare:", max_fare)
print("Average Fare:", mean_fare)
```

Task 13: Plot histogram of Age.

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

Task 14: Plot bar chart of Survival Count.

```python
import matplotlib.pyplot as plt


df['Survived_Label'] = df['Survived'].map({0: 'Not Survived', 1: 'Survived'})


counts = df['Survived_Label'].value_counts()


counts.plot(kind='bar', color=['red', 'green'])

plt.title("Survival Count")
plt.xlabel("Status")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=0)

plt.show()
```
